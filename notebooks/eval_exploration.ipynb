# LLM eval result exploration

Interactive analysis of RAGAS eval runs stored in `llm.eval_results` and cost
tracking in `llm.cost_log`. Requires the shared Postgres instance to be running.

```python
import pandas as pd
from src.utils.db import get_connection

conn = get_connection()
results = pd.read_sql("SELECT * FROM llm.eval_results ORDER BY evaluated_at DESC", conn)
results.head()
```

```python
metric_cols = ["faithfulness", "answer_relevancy", "context_precision", "context_recall"]
results.groupby("run_id")[metric_cols].mean()
```

```python
costs = pd.read_sql("SELECT run_type, SUM(cost_usd) AS total_usd FROM llm.cost_log GROUP BY run_type", conn)
costs
```
